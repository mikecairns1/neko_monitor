```python
from kafka import KafkaProducer, KafkaConsumer

producer = KafkaProducer(bootstrap_servers='192.168.2.9:9092')
```

```python
# 'cat_water' というトピックにメッセージを送る
producer.send('cat_water', b'Hey')
```

```python
# cat_waterに対してデータを取りに行く
consumer = KafkaConsumer(
    'cat_water', group_id='my_group', enable_auto_commit=True
    bootstrap_servers=['192.168.2.9:9092'])

# 最新データ
consumer.poll()
consumer.seek_to_end()

# メッセージの subscribe
for message in consumer:
    print("topic: %s message=%s" % (message.topic, message.value))
```

```python
from kafka import KafkaProducer
import requests

producer = KafkaProducer(bootstrap_servers='192.168.2.21:9092')

#producer.send('cat_water', b'Hey')

url = "http://192.168.2.30:8765/picture/1/current/"

try:
    r = requests.get(url)
except requests.exceptions.RequestException as e:
    #http error, connection error, timeout
    print("Request Error")

if r.status_code == 200:
    try:
        producer.send('cat_water', r.content)
    except:
        print("Error")
        
producer.flush()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from io import BytesIO

url = "http://192.168.2.30:8765/picture/1/current/"

try:
    r = requests.get(url)
except requests.exceptions.RequestException as e:
    #http error, connection error, timeout
    print("Request Error")

if r.status_code == 200:
    try:
        img = mpimg.imread(BytesIO(r.content), format='jpg')
        fig, ax = plt.subplots()
        ax.imshow(img)
        plt.show()
    except:
        print("Error")
```
